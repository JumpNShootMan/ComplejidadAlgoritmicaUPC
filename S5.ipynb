### Búsqueda por Costo Uniforme UCS

Ir de un nodo S a un nodo G al menor costo posible. Los vértices representan costos

```python
from random import *
import networkx as nx
import heapq as pq

def genGrafosPesos(n, m):
    matriz = [[]for i in range(n)]
    for j in range(m):
        v1 = randint(0,n-1)
        v2 = randint(0,n-1)
        w = randint(1,11)
        adyacentes = [x[0] for x in matriz[v1]]
        while(v2 in adyacentes or v2 == v1):
            v1 = randint(0,n-1)
            v2 = randint(0,n-1) 
        matriz[v1] += [(v2,w)]
    return matriz

def dibujarGrafoConPesos(lista):
    G = nx.DiGraph()
    for v1 in range(len(lista)):
        for v2 in lista[v1]:
            G.add_edge(v1, v2[0],weight=v2[1])
    nx.draw_circular(G, with_labels=True, font_weight='bold', node_size=1000)
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G,pos=nx.circular_layout(G),label_pos=0.3,edge_labels=labels, node_size=2000)
g = genGrafosPesos(10,20)
print(g)
dibujarGrafoConPesos(g)
```
